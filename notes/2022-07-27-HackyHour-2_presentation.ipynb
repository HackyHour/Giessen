# What are Jupyter Notebooks?

* Notebooks are documents produced by the Jupyter Notebook App
* contain both computer code (e.g. python) & rich text elements (paragraph, equations, figures, links, etc…)
* they are human-readable documents & executable documents
* great for developing code & communicating results

-> you can show text, figures and executable code in the same document!

# What are Jupyter Notebooks?

* Project Jupyter is a nonprofit organization
* to develop open-source software, open-standards, and services for interactive computing across dozens of programming languages
* supports execution environments in several dozen languages
* name **JUPYTER** similar to **JU**lia, **PYT**hon, **R**



Image source: https://ipython.org/ipython-doc/3/development/how_ipython_works.html

# Why are Jupyter Notebooks useful? 

* very transparent processes because code documentation can be implemented in a very clean way
* have the whole  “data science process” in one document
* IDE, presentation & educational tool combined
* execute code, see what happens, modify, and repeat in a kind of iterative conversation between user and data
* it is great for tutorials and to show how e.g. a package is used
